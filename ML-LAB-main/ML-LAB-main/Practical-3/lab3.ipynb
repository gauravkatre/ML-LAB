```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
print(df.info())
print(df.head())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df.describe()
```

```python
df.loc[5:10, 'sepal width (cm)']=np.nan
df.loc[1:5, 'sepal length (cm)']=np.nan
df.loc[10:15, 'petal length (cm)']=np.nan
df.loc[15:20, 'petal width (cm)']=np.nan
```

```python
df
```

```python
df['sepal width (cm)'] = df['sepal width (cm)'].fillna(df['sepal width (cm)'].mean())
df['sepal length (cm)'] = df['sepal length (cm)'].fillna(df['sepal length (cm)'].mean())
df['petal length (cm)'] = df['petal length (cm)'].fillna(df['petal length (cm)'].mean())
df['petal width (cm)'] = df['petal width (cm)'].fillna(df['petal width (cm)'].mean())
df
```

```python
sns.boxplot(data=df)
```
